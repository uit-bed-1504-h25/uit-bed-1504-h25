# For-l칮kker

Her er eksempel p친 en `for`-l칮kke som deriverer alle utrykkene i listen `expressions`:

#### Eksempel 1:
Vi har tidligere i kurset sett p친 ```sympy``` og derivasjon - dette kan vi ogs친 gj칮re ved hjelp av en ```for```-l칮kke, for 친 iterere gjennom en liste. 

````python
from IPython.display import Markdown, display
import sympy as sp

x,y=sp.symbols("x y")

expressions=[
    sp.root(x,y),
    x**2+5*sp.exp(x),
    x*sp.ln(x)
]

for i in expressions:
    display(Markdown(f"Den deriverte av ${sp.latex(i)}$ er ${sp.latex(sp.diff(i,x))}$"))
````

````python
for i in range(5,8):
    print(i)
    
for i in range(3):
    print(i)
````

Du kan ogs친 "pakke ut" elementer, dersom hvert element i `for`-setningen har et bestemt antall sub-elementer. La oss f칮rst lage en liste som inneholder en 췂tuple췂 med uttrykkene i `expressions` over, og den deriverte av disse:

#### Eksempel 3:

````python
#using list comprehension to generate a list with expression,derivative tuples:
expressions_w_derivatives=[ (i, sp.diff(i,x)) for i in expressions ]
expressions_w_derivatives
````

Vi kan n친 kj칮re en for-l칮kke som forventer at hvert element i listen er en *iterabel* (for eksempel liste eller tuple) med to elementer:

#### Eksempel 4

````python
for function,derivative in expressions_w_derivatives:
    display(Markdown(f"Den deriverte av ${sp.latex(function)}$ er ${sp.latex(derivative)}$"))
````

# While-l칮kker

En `while`-l칮kke er en l칮kke som fortsetter inntil en betingelse er oppfylt. I eksemplet under ser vi at det genereres nye tilfeldige aksjekurser s친 lenge kursen er under 125 kroner . 

Aksjen skal bevege seg tilfeldig. For 친 f친 til det bruker vi funksjonen `rand` fra `random`-modulen i numpy.  `np.random.rand()` gir et tilfeldig tall mellom null og 칠n, s친 ved 친 multiplisere `(np.random.rand()-0.5)` med 40, f친r vi et tilfeldig tall mellom -20 og 20. 

#### Eksempel 5:

````python
import numpy as np

dy, y = 0, 100

while y < 125:  
    y += dy
    dy = 40*(np.random.rand()-0.5) 
    print(y)
````

Legg spesielt merke til `y+=dy` over. Denne operasjonen legger `dy` til `y`, og tilsvarer alts친 `y=y+dy`. De aller fleste programmeringsspr친k st칮tter `+=`-operatoren.  

Legg ogs친 merke til at vi kan sette flere variabler samtidig, ved 친 skille b친de variablene og verdiene med like mange kommaer, `dy,y=0,100`

## Simulering av aksjekurs 游눷

I denne simuleringen antar vi at en aksjekurs starter p친 100 kr og beveger seg tilfeldig over tid.  
Vi observerer aksjen hvert sjette minutt, eller 칠n tidel (0.1) av en time. Tiden 칮ker dermed med 0.1 i hver periode. 

For 친 tegne opp stien til aksjekursen, lager vi f칮rst lister for x- og y-verdiene, og s친 legger vi til elementer i hver av listene med `append`-funksjonen til listeobjektene, f칮r listen plottes som en graf. Kj칮r koden flere ganger, for 친 se ulike simuleringer. 

#### Eksempel 6:

````python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Ellipse
from IPython.display import display, clear_output

# Oppretter en figur og akse med en st칮rrelse p친 20x10 tommer
fig, ax = plt.subplots(figsize=(20, 10))

# Setter begrensninger for y- og x-aksen
ax.set_ylim([0,250])            # Y-aksen g친r fra 0 til 250
ax.set_xlim([0,20])             # X-aksen g친r fra 0 til 20

# Initierer med en rekke verdier (x = 0, y = 100, dy = 0)
x, y, dy = 0 ,100 ,0

# Lager tomme lister som vil lagre x- og y-posisjoner gjennom l칮kka
ypath = []
xpath = []

# While-l칮kka kj칮rer s친 lenge x < 25, og for hvert steg 칮kes x med 0.1
while x < 25:
    x += 0.1            # 칒ker x med 0.1 for hvert steg
    y += dy             # Legger til dy til y-posisjonen (endrer y basert p친 dy)
    
    # Oppdaterer dy med en ny tilfeldig verdi som varierer mellom -20 og 20.
    # np.random.rand() gir en tilfeldig verdi mellom 0 og 1, 
    # s친 40 * (np.random.rand() - 0.5) gir et tilfeldig tall mellom -20 og 20.
    # Dette gir y-posisjonen en tilfeldig variasjon ved hvert steg.
    dy = 40 * (np.random.rand() - 0.5)

    # Legger til de nye x- og y-verdiene i listene xpath og ypath
    xpath.append(x)
    ypath.append(y)

# Plotter banen for y-verdiene som funksjon av x, med etiketten 'YARA'
ax.plot(xpath, ypath, label='YARA')

# Legger til en legende (forklaring) 칮verst til venstre p친 plottet
ax.legend(loc='upper left', frameon=False, fontsize=30)

# Viser plottet
plt.show()
````

# Dynamisk plott 游늴

Om vi 칮nsker 친 formidle kunnskap, er det av og til nyttig med dynamiske plott. Alts친 plott som endrer seg mens du ser p친 dem. Vi kan gj칮re det ved 친 rykke inn de to siste setningene fra Eksempel 6. 

Det er disse setningene som lager plottet. Ved 친 rykke dem inn, plottes figuren i hver iterasjon. 

### 游댳 Viktig 친 merke seg:
- Uten **`clear_output(wait=True)`**, ville vi sett **25 separate figurer** i stedet for 칠n dynamisk oppdatering. 
- I tillegg, s친 legger vi til at y-verdien ikke kan bli mindre enn $0$. Det gj칮r vi her **`while x < 25 and y >=0`**.

Kj칮r koden i Eksempel 7 og se hvordan den simulerte bevegelsen utvikler seg i sanntid! 

#### Eksempel 7:

````python
x, y, dy = 0, 100, 0

ypath = []
xpath = []

while x < 25 and y >=0:
    x += 0.1
    y += dy
    dy = 40 * (np.random.rand() - 0.5)

    xpath.append(x)
    ypath.append(y)
    
    # Ved 친 plassere plot-funksjonen inne i l칮kka, s칮rger vi for at den plottes ved hver iterasjon
    ax.plot(xpath, ypath, label='YARA')
    ax.legend(loc='upper left', frameon=False, fontsize=30)
    
    # Jupyter viser som standard bare siste utsagn, s친 vi m친 legge til dette
    # for 친 plotte p친 hver iterasjon
    display(fig)
    
    # S칮rger for at plottet og utdataene blir slettet hver gang:
    ax.cla()
    ax.set_ylim([0, 400])
    ax.set_xlim([0, 20])
    clear_output(wait=True)
````

# 游댝 Numerisk optimering  

Tidligere har vi funnet maksverdien av funksjoner ved 친 bruke **Sympy** til 친 derivere en funksjon, sette den lik null og l칮se for $x$.  

Men hva om det er vanskelig 친 finne den optimale l칮sningen analytisk? (Noe det ofte kan v칝re!) 

I slike tilfeller kan vi bruke **numerisk optimering**, der vi evaluerer objektfunksjonen (det vi 칮nsker 친 maksimere) for mange ulike verdier og velger den beste.  

### Hva gj칮r vi i dette eksemplet?  
Vi ser p친 en bedrift som selger $Q$ enheter av en vare til pris $p$, med en kostnad per enhet $C$.  
Bedriftens **fortjeneste** (profit) er gitt ved:  

$p \cdot Q - C \cdot Q$

Vi bruker numeriske metoder for 친 finne verdien av $Q$ som maksimerer fortjenesten.  

Videre, antar vi at ettersp칮rselfunksjonen (som bestemmer $Q$) er definert slik:

$100 - p$

Selv om vi i dette tilfellet enkelt kan finne maksimum analytisk med **Sympy**, kan vi bruke dette eksempelet for 친 sjekke at den numeriske l칮sningen er korrekt.  

Numerisk optimering er nyttig n친r:
- Funksjonen er **komplisert** og vanskelig 친 derivere.
- Vi har **diskrete verdier** der derivasjon ikke er praktisk.
- Vi 칮nsker 친 **visualisere hvordan l칮sningen varierer** over et bredt spekter av verdier.

La oss n친 implementere dette i Python i eksemplene under!

### Eksempel 11:

````python
# Definerer fortjenestefunksjonen, som avhenger av en ettersp칮rsel.
def profit(p, C):
    """Beregner profitt gitt pris p og kostnad per enhet C."""
    Q = demand(p)
    return p * Q - C * Q
````

Q bestemmes av en ettersp칮rselfunksjon.

#### Eksempel 12

````python
# Definerer ettersp칮rselfunksjonen
def demand(p):
    """Returnerer antall enheter solgt gitt en pris p."""
    return 100 - p
````

Vi kan n친 finne optimum numerisk, alts친 tallverdiene (optimal pris og hva maksimal profitt er ved denne optimale prisen.)

#### Eksempel 13:

````python
import numpy as np

def find_optimal_price(start, end, C):
    """Finner optimal pris i prisintervallet (start, end)"""
    max_profit = float('-inf')
    optimal_price = None
    n_samples = 1000
    
    # Itererer over pris-intervallet (start, end)
    for p in np.linspace(start, end, n_samples):
        current_profit = profit(p, C)
        if current_profit > max_profit:
            # Hvis 'profit' g친r over 'current profit', s친 lagrer vi den nye 'profit' og pris
            max_profit = current_profit
            optimal_price = p

    return optimal_price, max_profit


cost = 5
start_price = 0
end_price = 100

optimal_price, max_profit = find_optimal_price(start_price, end_price, cost)

print(f"Optimal pris: {optimal_price}")
print(f"Maksimal profitt ved optimal pris: {max_profit}")
````

Vi kan verifisere dette enda en gang, ved hjelp av ```SymPy```.

#### Eksempel 14:

````python
import sympy as sp

# Definerer variabelen p (pris)
p = sp.Symbol('p')

# Kostnad C er satt til 30
C = 5

# Ettersp칮rselsfunksjon
def demand(p):
    return 100 - p

def profit(p, C):
    """Beregner profitt gitt pris p og kostnad per enhet C."""
    Q = demand(p)
    return p * Q - C * Q

# Deriverer profittfunksjonen med hensyn til p
profit_derivative = sp.diff(profit(p, C), p)

# Finn den optimale p-verdien ved 친 l칮se ligningen profit_derivative = 0
optimal_p = sp.solve(profit_derivative, p)

# Viser den optimale prisen
optimal_p_value = round(optimal_p[0], 2)
optimal_p_value
````

# Monte Carlo-simulering 游둰勇 游 游늵

Monte Carlo-simulering er en metode for 친 estimere resultater ved hjelp av tilfeldige pr칮ver.  
Denne teknikken er spesielt nyttig n친r det er **usikkerhet** i dataene eller **mange mulige utfall**.  

### 游댳 Hvorfor bruke Monte Carlo-simulering?  
- Mange 칮konomiske prosesser, som **investeringer og l칮nnsvekst**, er usikre.  
- Deterministiske modeller fanger **ikke opp tilfeldigheter**, noe Monte Carlo-simulering gj칮r.  
- Vi kan bruke denne teknikken for 친 lage **sannsynlighetsfordelinger** og forutsi mulige utfall.  

I dette eksempelet ser vi p친 **inntektsutvikling** over tid, hvor inntekten hvert 친r avhenger av fjor친ret.  
- **Vi starter med en gitt inntekt** og antar en 친rlig vekstrate (칠n prosent).  
- **Hvert 친r legges en tilfeldig variasjon til**, som representerer usikkerhet.  
- Vi kj칮rer simuleringen mange ganger for 친 se hvordan inntekten utvikler seg over tid.  

Vi starter med 친 definere en funksjon som beregner neste 친rs inntekt, basert p친 fjor친ret og et tilfeldig bidrag: 

#### Eksempel 15

````python
import numpy as np

def calc_income(last_year_income, growth_rate, volatility):
    
    random_growth = np.random.normal(0, volatility)
    growth =  growth_rate + random_growth
    
    income = round(last_year_income * (1 + growth))
    
    return income

# Her antar vi at fjor친rets l칮nn er 600.000kr, med 5% forventet inntekts칮kning og 2% usikkerhet

print(f'Hvis fjor친rets l칮nn er 600.000 kr, s친 er 친rets l칮nn {calc_income(600000, 0.05, 0.02)}, hvis vi har 5% forventet l칮nns칮kning og 2% usikkerhet.')
````

#### **S친 lager vi en funksjon som lager en "inntektssti" (income path) for et gitt antall 친rs inntekt. Dette kalles en "sti"/"path" fordi dagens inntekt er avhengig av g친rsdagen. Dagens situasjon er alts친 avhengig av hvilken sti inntekten har fulgt i alle tidligere 친r.**

#### Eksempel 16

````python
def income_path(years, initial_income, growth_rate, volatility):
    
    incomes = [initial_income]
    
    #simulating a single income path:
    for year in range(years-1):

        #calculating income:
        income = calc_income(incomes[-1], growth_rate, volatility)

        #adding income for year to the path
        incomes.append(income)
        
    return incomes

#regner ut for fem 친r:

income_path(5,600000, 0.05, 0.02)
````

#### **Vi kan n친 lage en funksjon som simulerer et gitt antall inntektsstier gitt ved `num_simulations`:**

#### Eksempel 17:

````python
import pandas as pd

def simulate_income(years, initial_income, growth_rate, volatility, num_simulations):
    """Returnerer en 2D-array med simulerte inntekter for hvert 친r."""
    # Starter med en tom liste, hvor vi skal fylle p친 med tall
    all_incomes = []
    for _ in range(num_simulations):
        # Legger til "the path" til variabelesn 'all_incomes'
        incomes = income_path(years, initial_income, growth_rate, volatility)
        all_incomes.append(incomes)
    plt.figure(figsize=(10, 5))
        
    return np.array(all_incomes)

# Setter inn i pandas DataFrame for 친 f친 en visuelt, finere tabell

pd.DataFrame(simulate_income(5, 600000, 0.05, 0.02, 5))
````
