# ✏️ Oppgave 1 - Tilbud og etterspørsel
La oss se på et marked der etterspørselen ($D$) og tilbudet ($S$) for et produkt er gitt ved følgende funksjoner:

* **Etterspørsel:** $D(p) = 100 - 2p$
* **Tilbud:** $S(p) = 3p - 20$

Der $p$ er prisen på produktet i kroner, og $D(p)$ og $S(p)$ representerer henholdsvis mengden som etterspørres og tilbys ved prisen $p$.

**Finn likeveksprisen $p$, der tilbudet er lik etterspørselen**

````python
import sympy as sp
p = sp.symbols('p')

#Etterspørsel & tilbud
D = 100 - 2*p
S = 3*p - 20

#D = S, D-S=0

#likevekt
likevekt = sp.solve(D - S,p)
print(f'Likevekt når prisen {p} er lik {likevekt[0]}.')
````

# ✏️ Oppgave 4
I denne oppgaven skal vi se på Cobb-Douglas funksjonen, den ser slik ut
$$f(K, L) = A \cdot K^{\alpha} \cdot L^{\beta}$$
hvor,
- $f$ er total produksjon
- $K$ er mengden kapital (for eksempel maskiner, utstyr, bygninger)
- $L$ er mengden arbeidskraft (antall arbeidere eller arbeidstimer)
- $A$ er et mål på produktivitet
- $\alpha$ og $\beta$ er eksponenter som viser elastisiteten til henholdsvis kapital og arbeidskraft. De representerer hvor stor andel av produksjonen som kommer fra kapital og arbeidskraft.

Typisk er at $\alpha + \beta = 1$, som som indikerer konstante skalautbytter. Dette betyr at hvis du fordobler både kapital og arbeidskraft, vil produksjonen også fordobles.

## Oppgave 4.1

## Oppgave 4.3
Hvis vi holder $L$, $A$, $\alpha$ og $\beta$ konstant. Hvordan utvikler MPK seg, hvis $K$ øker? 

Velg selv en passende måte å presentere dette på. Beskriv hva dette betyr for den totale produksjonen til Cobb-Douglas funksjonen, $f(K, L)$.

Du bestemmer selv verdiene på $L$, $A$, $\alpha$ og $\beta$ (men sørg for $0 < \alpha < 1$, $0 < \beta < 1$ og $\alpha + \beta =1$).

````python
# 4.1
# Cobb- Douglas

def cobb_douglas(K, L, A=1, alpha=0.5, beta=0.5):
    return A*(K**alpha)*(L**beta)
````

````python
#4.2
import sympy as sp
K, L , A, alpha, beta = sp.symbols('K L  A alpha beta')

#Partiellderiverte
dY_dK = sp.diff(cobb_douglas(K, L, A, alpha, beta), K)
display(dY_dK)
````

Hvis vi kalles Cobb-Douglas funksjonen for $f(K, L)$, og så ser vi på den partiell-deriverte av Cobb-Douglas funksjonen, også kalt MLP, som vi fant over ser vi at:

$$MPK = \frac{AK^{\alpha}L^{\beta}\alpha}{K}$$

kan skrives som

$$MPK = \alpha \cdot \frac{f(K, L)}{K}$$

````python
# 4.3
import numpy as np
import matplotlib.pyplot as plt

def MPK(K, L=10, A=1, alpha=0.5, beta=0.5):
    return (A*alpha*K**(alpha)*L**(beta))/K

Kval = np.linspace(1, 200, 100)
MPKval = MPK(Kval)

plt.plot(Kval, MPKval, label='MPK')
plt.title('Utvkl av MPK når K øker')
plt.xlabel('Kapital (K)')
plt.ylabel('MPK')
plt.grid(True)
plt.legend()
plt.show()
````

### **Kommentarer til Oppgave 4.3**
Her har vi plottet $MPK$, som en funksjon av kapital $K$. At MPK, den partiell-deriverte av $f(K, L)$ (med hensyn på $K$), alltid er positiv betyr at produksjonen $f$ alltid vil øke når kapitalmengden $K$ øker. Gitt at vi holder alle andre variabler konstante, altså $L$, $A$, $\alpha$ og $\beta$. 

Selv om marginalproduktet av kapital avtar når $K$ øker (som plottet over viser), vil det aldri bli negativt. Skal vi tolke dette i en mer økonomisk forståelse av produksjon, kan man si at i en typisk produksjonsprosess vil mer kapital (som f.eks. maskiner, teknologi osv.) bidra til økt produksjon, men etter hvert vil denne økningen ha mindre og mindre effekt.
