# 九勇 Oppgave 1 - Tilbud og ettersp칮rsel
La oss se p친 et marked der ettersp칮rselen ($D$) og tilbudet ($S$) for et produkt er gitt ved f칮lgende funksjoner:

* **Ettersp칮rsel:** $D(p) = 100 - 2p$
* **Tilbud:** $S(p) = 3p - 20$

Der $p$ er prisen p친 produktet i kroner, og $D(p)$ og $S(p)$ representerer henholdsvis mengden som ettersp칮rres og tilbys ved prisen $p$.

**Finn likeveksprisen $p$, der tilbudet er lik ettersp칮rselen**

````python
# Vi skal finne der D(p) = S(p).
# Alts친 100 - 2p = 3p - 20

### L칮sningsforslag ###

import sympy as sp

# Definerer variabelen v친r p (pris)
p = sp.symbols('p')

# Definerer ettersp칮rsel- og tilbudsfunksjonene
D = 100 - 2*p
S = 3*p - 20

# L칮ser ligningen D(p) = S(p)
likevekt = sp.solve(D - S, p)

# Svaret
print(f'Det er likevekt n친r prisen {p} er lik {likevekt[0]}!')
````

# 九勇 Oppgave 2 - Visualisering av likevekt
I oppgave 1 fant du likevektsprisen $p$ der tilbud og ettersp칮rselen er like. N친 skal du lage en graf som viser:

1. Tilbudskurven $S(p)$
2. Ettersp칮rselskurven $D(p)$

**Hint:** Her kan du f친 bruk for b친de NumPy og Matplotlib.

````python
import numpy as np
import matplotlib.pyplot as plt

# Definer funksjonene for ettersp칮rsel og tilbud
def D(p):
    return 100 - 2*p

def S(p):
    return 3*p - 20

# Lage et utvalg av priser fra 0 til 50
p_values = np.linspace(0, 50, 500)

# Beregn ettersp칮rsel og tilbud
D_values = D(p_values)
S_values = S(p_values)

# Finn likevektspunktet (fra oppgave 1, der D = S)
# Vi finner at likevekt er p = 24
p_eq = 24
q_eq = D(p_eq)  # eller S(p_eq) - begge gir samme resultat

# Plotte kurvene
plt.plot(p_values, D_values, label="Ettersp칮rsel D(p) = 100 - 2p", color="blue")
plt.plot(p_values, S_values, label="Tilbud S(p) = 3p - 20", color="red")

# Markere likevektspunktet p친 plottet
plt.scatter(p_eq, q_eq, color="green", zorder=5)
plt.text(p_eq + 1, q_eq + 2, f'Likevekt: p={p_eq}, q={q_eq}', color="green")

# Akser og titler
plt.xlabel('Pris (p)')
plt.ylabel('Mengde (q)')
plt.title('Marked: Ettersp칮rsel og Tilbud')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()

# Vise grafen
plt.show()
````

# 九勇 Oppgave 4
I denne oppgaven skal vi se p친 Cobb-Douglas funksjonen, den ser slik ut
$$f(K, L) = A \cdot K^{\alpha} \cdot L^{\beta}$$
hvor,
- $f$ er total produksjon
- $K$ er mengden kapital (for eksempel maskiner, utstyr, bygninger)
- $L$ er mengden arbeidskraft (antall arbeidere eller arbeidstimer)
- $A$ er et m친l p친 produktivitet
- $\alpha$ og $\beta$ er eksponenter som viser elastisiteten til henholdsvis kapital og arbeidskraft. De representerer hvor stor andel av produksjonen som kommer fra kapital og arbeidskraft.

Typisk er at $\alpha + \beta = 1$, som som indikerer konstante skalautbytter. Dette betyr at hvis du fordobler b친de kapital og arbeidskraft, vil produksjonen ogs친 fordobles.

## Oppgave 4.3
Hvis vi holder $L$, $A$, $\alpha$ og $\beta$ konstant. Hvordan utvikler MPK seg, hvis $K$ 칮ker? 

Velg selv en passende m친te 친 presentere dette p친. Beskriv hva dette betyr for den totale produksjonen til Cobb-Douglas funksjonen, $f(K, L)$.

Du bestemmer selv verdiene p친 $L$, $A$, $\alpha$ og $\beta$ (men s칮rg for $0 < \alpha < 1$, $0 < \beta < 1$ og $\alpha + \beta =1$).

````python
# Oppgave 4.1 

# Definerer Cobb-Douglas-produksjonsfunksjonen, etter oppgavetekstens beskrivelse
def cobb_douglas(K, L, A=1, alpha=0.5, beta=0.5):
    return A * (K ** alpha) * (L ** beta)
````

````python
# Oppgave 4.2 
import sympy as sp

# Definer symboler for kapital (K), arbeidskraft (L), og konstanter A, alpha, beta
K, L = sp.symbols('K L')
A, alpha, beta = sp.symbols('A alpha beta')

# Finn partielle deriverte med hensyn til kapital (K)
dY_dK = sp.diff(cobb_douglas(K, L, A, alpha, beta), K)

print("Den partielle deriverte med hensyn p친 K er:")
display(dY_dK)
````

Hvis vi kalles Cobb-Douglas funksjonen for $f(K, L)$, og s친 ser vi p친 den partiell-deriverte av Cobb-Douglas funksjonen, ogs친 kalt MLP, som vi fant over ser vi at:

$$MPK = \frac{AK^{\alpha}L^{\beta}\alpha}{K}$$

kan skrives som

$$MPK = \alpha \cdot \frac{f(K, L)}{K}$$

````python
# Oppgave 4.3 

import numpy as np
import matplotlib.pyplot as plt

# Definerer en funksjon for MPK basert p친 derivert av Cobb-Douglas
def MPK(K, L=10, A=1, alpha=0.5, beta=0.5):
    return (A * alpha * K**(alpha) * L**beta)/K

# Verdier for K
K_values = np.linspace(1, 200, 100)

# Beregn MPK for hver verdi av K
MPK_values = MPK(K_values)

# Plotter MPK som funksjon av K
plt.plot(K_values, MPK_values, label='MPK (marginalprodukt av kapital)')
plt.title('Utviklingen av MPK n친r K 칮ker')
plt.xlabel('Kapital (K)')
plt.ylabel('MPK')
plt.grid(True)
plt.legend()
plt.show()
````

### **Kommentarer til Oppgave 4.3**
Her har vi plottet $MPK$, som en funksjon av kapital $K$. At MPK, den partiell-deriverte av $f(K, L)$ (med hensyn p친 $K$), alltid er positiv betyr at produksjonen $f$ alltid vil 칮ke n친r kapitalmengden $K$ 칮ker. Gitt at vi holder alle andre variabler konstante, alts친 $L$, $A$, $\alpha$ og $\beta$. 

Selv om marginalproduktet av kapital avtar n친r $K$ 칮ker (som plottet over viser), vil det aldri bli negativt.
