# Oppgave 1a
Lag et scatter-plot som viser forholdet mellom antall timer brukt på å studere (```Hours_Studied```) og eksamensresultatet (```Exam_Score```). Bruk en farge for å indikere kjønn (```Gender```) i plottet, hvor for eksempel blå representerer "Male" og rød representerer "Female".

Hva kan vi si om sammenhengen mellom antall timer studert og eksamensresultater? Ser det ut til å være en forskjell mellom kjønnene?

````python
import pandas as pd

#df = pd.read_csv('/din_path/StudentPerformanceFactors.csv') 
````

````python
url = 'https://raw.githubusercontent.com/uit-bed-1504-h25/uit-bed-1504-h25/main/data/StudentPerformanceFactors.csv'
df = pd.read_csv(url)
df.head()
````

````python
import matplotlib.pyplot as plt

#Lager farger basert på kjønn
colors = df['Gender'].map({'Male': 'blue', 'Female': 'red'})

#Scatter-plot av antall timer studert vs. eksamensresultat
plt.figure(figsize=(10, 6))
plt.scatter(df['Hours_Studied'], df['Exam_Score'], c=colors, alpha = 0.7)

# Tittel og akse-navn
plt.title('Studietimer vs Eksamensresultater etter kjønn')
plt.xlabel('Ant timer studert per uke')
plt.ylabel('Eksamensresultat')

# Legg til en legend
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Male'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='Female')])
# Vise plottet
plt.grid(True)
plt.show()
````

# Oppgave 1b
Lag et histogram som viser fordelingen av antall timer søvn per natt (```Sleep_Hours```). Del dataene inn i intervaller på én time.

Hva er gjennomsnittlig antall timer med søvn per natt blant studentene? Virker det som om søvnmengden har en innvirkning på eksamensresultatene?

````python
plt.figure(figsize=(10, 6))
plt.hist(df['Sleep_Hours'], bins = range(5, 10), alpha = 0.7, edgecolor = 'black')
plt.xlabel('Ant. timer søvn per natt')
plt.grid(axis='y')
plt.show()
````

# Oppgave 1c
Lag et linjediagram som viser hvordan tilstedeprosenten (```Attendance```) varierer med eksamensresultatene (```Exam_Score```). Sorter dataene etter tilstedeprosenten for en tydelig trend.

Hva ser du når du plotter tilstedeværelse og eksamensresultater? Virker det som om høyere fravær fører til dårligere eksamensresultater?

````python
df_sorted = df.sort_values('Attendance')

plt.figure(figsize = (10,6))
plt.plot(df_sorted['Attendance'], df_sorted['Exam_Score'], marker = 'o')

plt.xlabel('Tilstedeprosent')
plt.ylabel('Eksamensresultat')
plt.grid(True)
plt.show()
````

# Oppgave 1d
Lag en bar-plot som viser gjennomsnittlig eksamensresultat (```Exam_Score```) for de ulike nivåene av foreldreinvolvering (```Parental_Involvement```).

Hvilket nivå av foreldreinvolvering ser ut til å korrelere med høyere eksamensresultater?

Hint:
* ```mean()``` funksjonen kan brukes for å finne gjennomsnitt.
* Kan hjelpe å lage følgende variabel:
```python
avg_scores_by_involvement = df.groupby('Parental_Involvement')['Exam_Score'].mean()
```

````python
# Beregn gjennomsnittlig eksamensresultat for hver kategori av foreldreinvolvering
avg_scores_by_involvement = df.groupby('Parental_Involvement')['Exam_Score'].mean()

# Bar-plot av gjennomsnittlige eksamensresultater for hver kategori
plt.figure(figsize=(10, 6))
avg_scores_by_involvement.plot(kind='bar', color=['skyblue', 'orange', 'green'], edgecolor='black')

# Tittel og akse-navn
plt.title('Gjennomsnittlig eksamensresultat vs Foreldreinvolvering')
plt.xlabel('Foreldreinvolvering')
plt.ylabel('Gjennomsnittlig eksamensresultat')

plt.grid(True)
plt.show()
````
